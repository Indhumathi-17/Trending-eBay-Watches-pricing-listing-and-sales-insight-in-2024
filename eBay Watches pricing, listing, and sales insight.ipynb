```python
#for my analysis the subtitle column is not needed and it has the most null values, so i drop the whole column 
df.drop('subTitle',axis=1,inplace = True)
```

```python
df
```

```python
df.isnull().sum()
```

```python
#in the sold column I replace all null values with '0' because I take the null as still not sold which gives the clear view of the availability of the product
df['sold'].fillna(0, inplace=True)
```

```python
import datetime
```

```python
#for column where date time is null, i fill it with current date time for my convient
current_dt = datetime.datetime.now()
df ['lastUpdated'].fillna(current_dt,inplace = True)
```

```python
#for the itemlocation and type, i fill it with 'ffill' because it has more similar and often values
df['itemLocation'].fillna(method = 'ffill', inplace = True)
df['type'].fillna(method = 'ffill', inplace = True)

#for availabletext i fill the null values with avalilability because those which has 0 in sold will be meaning full if the watches are 'available' rather than filled with 'sold'
df['availableText'].fillna('available', inplace = True)
```

```python
#now I cleaned all the null values and lets go some more deep
df.isnull().sum()
```

```python
df
```

```python
#in this 'type' column we have mostly spelling mistakes and same words with different spells, for my convienent and not to skewed the data more, I do some changes based on the value counts (on my own) 
df['type'].unique()
```

```python
#it mostly shows that the wrist watches were the most sold items which is like they mention all as wrist watches rather than give the types
df['type'].value_counts()
```

```python
#I do some changes (based on my own insight) to get little interesting analysis, i give some types for each one and make it little less skewed
rename = {'Wristwatches':'Digital watch','Watches':'Analog watch','Wrist watch':'Digital watch', 'Value':'Unisex', 'Women':'Unisex','Men':'Unisex'}
```

```python
df['type'] = df['type'].replace(rename)
```

```python
#now we can do little elaborate analysis rather than having same single 'wristwatch' as type
df['type'].value_counts()
```

```python
df['seller'].unique()
```

```python
#here some have misspells and some has same name with different spellings so i clean it all with renaming

rename_1 = {'sigmatime':'Sigmatime','Times On Spot':'Spot On Times',
            'spot-on-times':'Spot On Times','directsourcegoldndiamond':'Direct Source G&D'}
```

```python
df['seller'] = df['seller'].replace(rename_1)
```

```python
df['seller'].unique()
```

```python
#changing date time to a single format and remove time for my convience

df['lastUpdated'] = pd.to_datetime(df['lastUpdated'])
df['Last Update'] = df['lastUpdated'].dt.strftime('%b %d, %Y')
df.drop(columns=['lastUpdated'], inplace = True)
```

```python
df.columns
```

```python
# To get better accessability and understanding i rename the columns
rename_columns = {'availableText':'Availability', 'itemLocation':'Location',
                  'itemNumber':'Item Number','priceWithCurrency': 'Price',
                  'title': 'Item Names','type':'Types of Watches','sold':'Number of Sales'
                  }
```

```python
df.rename(columns = rename_columns, inplace = True)
```
